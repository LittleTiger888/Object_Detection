```python
import tensorflow as tf

# List all available physical devices, specifically GPUs
gpu_devices = tf.config.list_physical_devices('GPU')

if gpu_devices:
    print(f"TensorFlow found the following GPU(s): {gpu_devices}")
    print(f"Number of GPUs available: {len(gpu_devices)}")
else:
    print("TensorFlow did not find a GPU. Code will run on CPU.")

# Optional: Print the name of the default GPU device
if tf.test.gpu_device_name():
    print('Default GPU Device: {}'.format(tf.test.gpu_device_name()))
```
